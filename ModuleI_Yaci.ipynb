2.2 Probability Distribution

Probability distribution -- is the mathematical function that gives the probabilities of occurrence of different possible outcomes for an experiment.

Conditional probability -- the likelihood of an event or outcome occurring, based on the occurrence of a previous event or outcome.

Random variable is -- a measurable function defined on a prob-
ability space that maps from the sample space to the real numbers.

A random variable X is continuous if possible values comprise either a single
interval on the number line or a union of disjoint intervals.

Normal distributions are often used in the natural and social sciences to
represent real-valued random variables whose distributions are not known.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y = 1/(np.sqrt(2*np.pi)) * np.exp(-(x**2)/2)
plt.plot(x, y)
plt.title('Standard Normal (z) Curve')
plt.xlabel('Z Score')
plt.ylabel('Probability Density')
plt.show()
```

2.3 Independent Variables and Random Samples

Joint probability -- the probability of two or more events
happening together. 

Probability mass function (pmf) of a single discrete random variable X
can be extended to two variables X ,Y for describing how much probability
mass is placed on each possible pair of values (x,y).

```python
#Joint probability
import numpy as np
import matplotlib.pyplot as plt

joint_prob = np.array([[0.1, 0.2, 0.1],
                       [0.2, 0.3, 0.1],
                       [0.1, 0.1, 0.1]])
x = np.arange(joint_prob.shape[0])
y = np.arange(joint_prob.shape[1])
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots()
im = ax.imshow(joint_prob, cmap='YlOrBr')
ax.set_xticks(x)
ax.set_yticks(y)
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_yticklabels(['X', 'Y', 'Z'])
ax.set_xlabel('Variable 1')
ax.set_ylabel('Variable 2')
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Joint Probability', rotation=-90, va="bottom")
plt.show()
```

```python
#Independent Random Variables
import numpy as np
import matplotlib.pyplot as plt

n = 100
mean_x, std_x = 0, 1
mean_y, std_y = 0, 2
X = np.random.normal(mean_x, std_x, n)
Y = np.random.normal(mean_y, std_y, n)
fig, ax = plt.subplots()
ax.scatter(X, Y, s=2, alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()

```

2.4 Maximum Likelihood Estimation

Maximum likelihood estimation (MLE) is an effective approach of estimating
the parameters of a probability distribution through maximizing a likelihood
function.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
mean_true, std_true = 0, 1
X = np.random.normal(mean_true, std_true, n)
def likelihood(X, mean, std):
    return np.prod(1 / (np.sqrt(2 * np.pi) * std) * np.exp(-((X - mean) ** 2) / (2 * std ** 2)))
mean_range = np.linspace(-3, 3, 100)
likelihoods = [likelihood(X, mean, std_true) for mean in mean_range]
mean_mle = mean_range[np.argmax(likelihoods)]
ax.plot(mean_range, likelihoods)
ax.axvline(mean_mle, color='r', linestyle='--')
ax.set_xlabel('Mean')
ax.set_ylabel('Likelihood')
plt.show()

```

```python
#Using Python to find the derivative

from sympy import * 
x = Symbol('x')
f = 2*x**4+3
print("Expression : {}".format(f))
f_prime = f.diff(x)
f_prime
```

```python
from sympy import * 
x = Symbol('x')
f = cos(x)+sin(x)
print("expression = ", f)
f_prime = f.diff(x)
f_prime
```

```python
#partial derivative
from sympy import *
x, y = symbols('x y')
f = x**2 + 2 * y + y**3
print("Expression : {}".format(f))
f_prime = Derivative(f, y)
print("Value of the derivative : {}".format(f_prime.doit()))
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.misc import derivative

def func(x):
  return 2*x**3+x**2+5*x

def deri(x):
  return derivative(func, x)

x = np.linspace(0,5)

plt.plot(x,deri(x),label="derivative")
plt.plot(x,func(x),label="expression")
plt.legend()
plt.show()
```
